Q4. What is DataFrame in pandas and how is it different from pandas.series? Explain with an example.

```python
'''
DataFrame:
In pandas, a DataFrame is a two-dimensional labeled data structure that can store data of various types 
(numeric, string, boolean, etc.) in a tabular form with rows and columns. It is similar to a spreadsheet 
or a SQL table. Each column of a DataFrame is a pandas Series object.
'''
data={
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
    'Enrollment': [101, 102, 103, 104, 105],
    'Grade': [87, 92, 78, 85, 90]
}

df=pd.DataFrame(data)
print("DataFrame:")
print(df)
print(type(df))
print()

'''
Series:
On the other hand, a pandas Series is a one-dimensional labeled array capable of holding data of any type 
(numeric, string, boolean, etc.). It is similar to a column in a spreadsheet or a single column in a SQL 
table.
'''
s=pd.Series(data['Grade'])
print("Series:")
print(s)
print(type(s))

```

Q7. Create a DataFrame using multiple Series. Explain with an example.

```python

#creating a series for name
data_name=['Alice', 'Bob', 'Charlie', 'David', 'Emily']
s1=pd.Series(data_name)

#creating a series for enrollment
data_enrollment=[101, 102, 103, 104, 105]
s2=pd.Series(data_enrollment)

#creating a series for grade
data_grade=[87, 92, 78, 85, 90]
s3=pd.Series(data_grade)

#combining all the series to make a dataframe
df1=pd.DataFrame({"Name":s1,"Enrollment":s2,"Grade":s3})

#printing the dataframe
print(df1)

```

Q3. What is App routing in Flask? Why do we use app routes?

```python
'''
-In Flask, app routing refers to the process of defining URL routes that map to specific functions in a Flask application. 
 App routing is a key feature of Flask that enables developers to create dynamic web applications that can respond to 
 different user requests.

-When a user requests a particular URL in a Flask application, the Flask routing system matches the URL to a specific function 
 that is responsible for handling the request. This function is known as a view function, and it is responsible for generating 
 the content that is returned to the user's browser.

-For example, suppose you have a Flask application that has a route defined for the URL "/hello". In that case, when a user 
 navigates to that URL in their browser, Flask's routing system will match the URL to the view function that is associated with 
 that route. The view function can then generate a response that is returned to the user's browser.

-We use app routes in Flask to define the URLs that users can access in our application and map them to specific view functions. 
 This allows us to create dynamic web pages that can display different content depending on the URL that the user requests. App 
 routes also enable us to organize our code and make it easier to maintain and update our Flask applications. Additionally, app 
 routes allow us to handle different types of HTTP requests, such as GET, POST, PUT, and DELETE, which are used to perform various 
 actions on web resources.

'''
```

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
from flask import Flask
app=Flask(__name__)

@app.route("/welcome")
def welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

@app.route("/")
def details():
    return "<h1>Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

```python
"""
url_for() function is used to build URL to the specific function dynamically. 
by this we can avoid hard-coding the URLs into the templates by dynamically building them using this

# """
from flask import Flask,redirect, url_for  
   
app = Flask(__name__)  
  
@app.route('/admin')  
def admin():  
    return 'admin'  
  
@app.route('/librarion')  
def librarion():  
    return 'librarion'  
  
@app.route('/student')  
def student():  
    return 'student'  
  
@app.route('/user/<name>')  
def user(name):  
    if name == 'admin':  
        #if we add user/admin then this will redirect to admin automatically 
        return redirect(url_for('admin'))  
    if name == 'librarion':  
        return redirect(url_for('librarion'))  
    if name == 'student':  
        return redirect(url_for('student'))  
if __name__ =='__main__':  
    app.run(host="0.0.0.0")  

#so with the help url_for function we bind users weblink to our specific functions automatically
#whenever we enter '127.0.0.1:5000/user/admin' it will be redirected to '127.0.0.1:5000/admin' and same with others 
```
